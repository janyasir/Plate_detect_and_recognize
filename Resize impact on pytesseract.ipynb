# Aim
We want to test how the pytesseract performance improves/degrades by varying the size of the image.
So we test on 3 sample images 

```python
import cv2
import pytesseract
import numpy as np
import matplotlib.pyplot as plt
import glob

import time
import math
from termcolor import colored
tess_config = "-c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789 --psm 6"
```

```python
def image_pytesseract_OCR(img):
    final_string = resized_image_string = pytesseract.image_to_string(img, config=tess_config, lang="eng")
    return final_string
        
```

```python
img_folder_path = ('../plate_samples_256_64/*.jpg')
fnames = glob.glob(img_folder_path)


fnames.sort()
#print(fnames)

cols = 3
rows = math.ceil(len(fnames)/cols)

plt.figure(figsize=(30,rows*5))
for fcnt, f in enumerate(fnames):
    img = plt.imread(f)
    plt.subplot(rows,cols,fcnt+1)    
    plt.imshow(img)
    title = str(img.shape)
    plt.title(title, fontsize=25 )
```

```python
def get_allsize_pytesseract_strings(test_image, maxw = 10, maxh = 5):
    maxw = 10
    maxh = 5

    sc = 64

    str_list = []
    size_list =[]
    test_roi_list = []
    

    #plt.figure(figsize = (30,30))
    imcnt = 1
    for resh in range(0,maxh):
        for resw in range(maxh+3,maxw):

            #print(imcnt)
            ww = sc + (resw*sc)
            hh = sc + (resh*sc)
            resized_test_image = cv2.resize(test_image, (ww,hh),interpolation = cv2.INTER_AREA)
            resized_image_string = image_pytesseract_OCR(resized_test_image.astype(np.uint8))
            titlestr = str(f'{ww}x{hh} : {resized_image_string}')

            size_list.append(str(f'{ww}x{hh}'))
            str_list.append(resized_image_string)

            #print(f'Detected strings : {titlestr}')


            imcnt = imcnt + 1
            
    return str_list, size_list
```

```python
start_time  = time.time()

total_img_cnt = 0

for fn in fnames:
    test_image = plt.imread(fn)
    
    test_image_string = image_pytesseract_OCR(test_image.astype(np.uint8))
    total_img_cnt = total_img_cnt + 1
    
    print(colored(f'Original img pred string : {test_image_string}','red'))    

    plt.figure()
    plt.imshow(test_image)
    plt.show()
    
    maxw = 10
    maxh = 5
    
    str_list, size_list = get_allsize_pytesseract_strings(test_image, maxw = maxw, maxh = maxh)
    total_img_cnt = total_img_cnt + len(str_list)
    print('String list of various resizes')
    #print(size_list)
    print(str_list)
    
        
        
    
    print('**************************************')
end_time  = time.time()
```

```python
print(f'Time taken : {total_img_cnt+1} images : {end_time - start_time}')
print(f'Average Time taken : 1 image : {(end_time - start_time)/(total_img_cnt+1)}')
```
